```python
import pandas as pd
import numpy as np

df = pd.read_csv('patient_survival.csv')
print(df)
```

```python
statistics = df.describe()
print(statistics)
```

```python
numerical_features = ['h1_spo2_min', 'h1_sysbp_max', 'h1_sysbp_min', 'h1_sysbp_noninvasive_max', 'h1_sysbp_noninvasive_min',
                      'd1_glucose_max', 'd1_glucose_min', 'd1_potassium_max', 'd1_potassium_min',
                      'apache_4a_hospital_death_prob', 'apache_4a_icu_death_prob']
```

```python
# Tính mean của các Numerical Features
mean_values = df[numerical_features].mean()
print("Mean values:")
print(mean_values)
```

```python
# Tính median của các Numerical Features
median_values = df[numerical_features].median()
print("\nMedian values:")
print(median_values)
```

```python
# Tìm giá trị nhỏ nhất của các Numerical Features
min_values = df[numerical_features].min()
print("\nMin values:")
print(min_values)
```

```python
# Tìm giá trị lớn nhất của các Numerical Features
max_values = df[numerical_features].max()
print("\nMax values:")
print(max_values)
```

```python
# Tính phân phối của các Numerical Features
distribution = df[numerical_features].describe()
print("\nDistribution:")
print(distribution)
```

```python
import matplotlib.pyplot as plt
# Vẽ histogram cho các Numerical Features
for feature in numerical_features:
    plt.figure()
    df[feature].plot(kind='hist', bins=20)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.show()
```

```python
# Vẽ box plot cho các Numerical Features
plt.figure(figsize=(15, 12))
df[numerical_features].plot(kind='box')
plt.title('Box Plot of Numerical Features')
plt.xlabel('Features')
plt.ylabel('Value')
plt.show()
```

```python
# Vẽ scatter plot cho cặp Numerical Features
plt.figure()
plt.scatter(df['apache_4a_hospital_death_prob'], df['apache_4a_icu_death_prob'])
plt.title('Scatter Plot of apache_4a_hospital_death_prob and apache_4a_icu_death_prob')
plt.xlabel('apache_4a_hospital_death_prob')
plt.ylabel('apache_4a_icu_death_prob')

plt.show()
```

```python
import seaborn as sns
# Chọn các đặc trưng và biến mục tiêu
features = ["h1_spo2_min", "h1_sysbp_max", "h1_sysbp_min", "d1_glucose_max", "d1_glucose_min"]
target = "hospital_death"

# Tính ma trận tương quan
correlation_matrix = df[features + [target]].corr()

# Vẽ heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
